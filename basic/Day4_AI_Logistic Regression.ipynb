# Logistic Regression Classification
 Logistic Regression ‡§≠‡§®‡•á‡§ï‡•ã ‡§ï‡•á ‡§π‡•ã?
üëâ ‡§Ø‡•ã ‡§è‡§ï‡§¶‡§Æ‡•à popular classification algorithm ‡§π‡•ã
üëâ ‡§Ø‡•ã continuous output ‡§¶‡§ø‡§Å‡§¶‡•à‡§® (‡§ú‡§∏‡•ç‡§§‡•ã Linear Regression)
üëâ ‡§¨‡§∞‡•Å ‡§Ø‡§∏‡§≤‡•á ‡§≠‡§®‡•ç‡§õ:

Yes ‡§ï‡§ø No

Pass ‡§ï‡§ø Fail

1 ‡§ï‡§ø 0

Spam ‡§ï‡§ø Not Spam

Yeslay Sigmoid function ‡§™‡•ç‡§∞‡§Ø‡•ã‡§ó ‡§ó‡§∞‡•ç‡§õ
 Sigmoid Function
‡§Ø‡•ã ‡§è‡§ï mathematical formula ‡§π‡•ã, ‡§ú‡§∏‡§≤‡•á ‡§ï‡•Å‡§®‡•à ‡§™‡§®‡§ø number ‡§≤‡§æ‡§à 0 ‡§∞ 1 ‡§¨‡•Ä‡§ö‡§Æ‡§æ convert ‡§ó‡§∞‡•ç‡§õ

Formula:

sigmoid(x) = 1 / (1 + e^(-x))
üëâ ‡§Ø‡•ã function ‡§≤‡•á probability ‡§®‡§ø‡§ï‡§æ‡§≤‡•ç‡§õ
üëâ ‡§ú‡§∏‡•ç‡§§‡•ã: "‡§§‡§ø‡§Æ‡•Ä‡§≤‡•á 5 ‡§ò‡§£‡•ç‡§ü‡§æ ‡§™‡§¢‡•ç‡§Ø‡•å ‡§≠‡§®‡•á ‡§™‡§æ‡§∏ ‡§π‡•Å‡§®‡•á chance 82%"



```python
# Python ‡§¨‡§æ‡§ü Sigmoid Function plot ‡§ó‡§∞‡•å‡§Å

import numpy as np 
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 /(1 + np.exp(-x))

#

x = np.linspace(-10, 10, 100) # x ‡§ï‡•ã ‡§≤‡§æ‡§ó‡§ø ‚àí10 ‡§¶‡•á‡§ñ‡§ø 10 ‡§∏‡§Æ‡•ç‡§Æ‡§ï‡§æ 100 ‡§µ‡§ü‡§æ number ‡§¨‡§®‡§æ‡§á‡§è‡§ï‡•ã ‡§π‡•ã

# linspace ‡§≤‡•á continuous range ‡§¨‡§®‡§æ‡§â‡§Å‡§õ
y = sigmoid(x)

plt.plot(x, y, color='green')
plt.title("Sigmoid Function")
plt.xlabel("Input")
plt.ylabel("probability")
plt.grid(True)
plt.show()

'''‡§ú‡§¨ x ‡§ß‡•á‡§∞‡•à negative ‡§π‡•Å‡§®‡•ç‡§õ ‚áí sigmoid almost 0 ‡§π‡•Å‡§®‡•ç‡§õ

‡§ú‡§¨ x positive ‡§ú‡§æ‡§®‡•ç‡§õ ‚áí sigmoid 1 ‡§§‡§ø‡§∞ ‡§™‡•Å‡§ó‡•ç‡§õ

‡§¨‡•Ä‡§ö‡§Æ‡§æ x = 0 ‡§Æ‡§æ sigmoid value ‡§π‡•Å‡§®‡•ç‡§õ 0.5'''
```

# Logistic Regression ‡§ï‡•ã Simple Example
‡§§‡§ø‡§Æ‡•Ä‡§≤‡•á 1 ‡§¶‡•á‡§ñ‡§ø 7 ‡§ò‡§£‡•ç‡§ü‡§æ ‡§™‡§¢‡•ç‡§®‡•á ‡§°‡•á‡§ü‡§æ ‡§õ‡•§
la aaba  ‡§è‡§â‡§ü‡§æ model ‡§¨‡§®‡§æ‡§Ø‡•å: ‡§ï‡§§‡§ø ‡§ò‡§£‡•ç‡§ü‡§æ ‡§™‡§¢‡•á‡§™‡§õ‡§ø ‡§™‡§æ‡§∏ ‡§π‡•Å‡§®‡•á ‡§∏‡§Æ‡•ç‡§≠‡§æ‡§µ‡§®‡§æ ‡§õ?

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# study hours ( input)
X = np.array([1,2,3,4,5,6,7]).reshape(-1, 1)

# labels : 0 = fail , 1 = pass
y = np.array([0,0,0,1,1,1,1])

# train the model
model = LogisticRegression()
model.fit(X, y)

# predict new data input

X_test = np.array([2,5,6]).reshape(-1, 1)
predictions = model.predict(X_test)
print("Predictions", predictions)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

y_true = [0, 1, 1, 0, 1]      # Actual labels
y_pred = [0, 1, 0, 0, 1]      # Predicted labels

print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
print("Accuracy:", accuracy_score(y_true, y_pred))
print("Precision:", precision_score(y_true, y_pred))
print("Recall:", recall_score(y_true, y_pred))
```

Q: Logistic Regression ‡§≤‡•á continuous value predict ‡§ó‡§∞‡•ç‡§õ ‡§ï‡§ø discrete class?

üëâ A:

Logistic Regression ‡§≤‡•á probability predict ‡§ó‡§∞‡•ç‡§õ, ‡§∞ ‡§§‡•ç‡§Ø‡§∏‡§≤‡§æ‡§à threshold ‡§ó‡§∞‡•á‡§∞ class ‡§Æ‡§æ ‡§¨‡§¶‡•ç‡§≤‡§ø‡§®‡•ç‡§õ (‡§ú‡§∏‡•ç‡§§‡•à 0.5 ‡§≠‡§®‡•ç‡§¶‡§æ ‡§Æ‡§æ‡§•‡§ø ‡§≠‡§®‡•á 1)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Input data: study hours
X = np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1, 1)

# Labels: 0 = fail, 1 = pass
y = np.array([0, 0, 0, 1, 1, 1, 1])

# Train model
model = LogisticRegression()
model.fit(X, y)

# Plotting decision boundary
x_range = np.linspace(0, 8, 300).reshape(-1, 1)  # Smooth curve input
probabilities = model.predict_proba(x_range)[:, 1]  # Probability of passing (class 1)

plt.figure(figsize=(8, 5))
plt.plot(x_range, probabilities, color='green', label='Pass Probability')
plt.scatter(X, y, color='blue', label='Training Data')

# Predicted points
X_test = np.array([2, 5, 6]).reshape(-1, 1)
predictions = model.predict(X_test)
plt.scatter(X_test, predictions, color='red', marker='x', s=100, label='Test Predictions')

# Styling
plt.title("Logistic Regression: Pass/Fail Based on Study Hours")
plt.xlabel("Hours Studied")
plt.ylabel("Probability of Passing")
plt.legend()
plt.grid(True)
plt.ylim(-0.1, 1.1)
plt.show()

# Console prediction output
print("Predictions:", predictions)
```
