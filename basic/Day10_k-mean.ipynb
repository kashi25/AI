```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Dummy data: income vs spending score
X = np.array([
    [15, 39],
    [16, 81],
    [17, 6],
    [18, 77],
    [19, 40],
    [20, 76],
    [21, 6],
    [22, 77],
    [23, 40],
    [24, 75]
])

# Apply KMeans
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids')
plt.title("Customer Clustering (KMeans)")
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Fruit dataset: [weight (grams), days_to_rot]
X = np.array([
    [120, 2],    # banana
    [135, 3],
    [250, 6],    # mango
    [260, 5],
    [400, 10],   # watermelon
    [420, 12],
    [130, 2],    # banana
    [245, 5],    # mango
    [410, 11],   # watermelon
])

# Clustering
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
labels = kmeans.predict(X)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Accent')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, label='Centroids')
plt.xlabel("Weight (grams)")
plt.ylabel("Days to Rot")
plt.title("Fruit Clustering using K-Means")
plt.legend()
plt.grid(True)
plt.show()
```
