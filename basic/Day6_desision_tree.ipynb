```python
import pandas as pd

# Dummy dataset of fruits
data = {
    "Weight": [100, 150, 120, 80, 200, 130, 90],
    "Size": [7, 8, 7, 6, 9, 8, 6],
    "Color_Score": [8.5, 9.0, 7.5, 6.0, 9.5, 8.0, 6.5],
    "Quality": ["Good", "Good", "Good", "Bad", "Good", "Good", "Bad"]
}

df = pd.DataFrame(data)
print(df)
```

```python
# lets' create a decision 
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

 # split features and target
X = df[["Weight" ,"Size" , "Color_Score"]]
y = df["Quality"]

# now time to train the model
tree = DecisionTreeClassifier(max_depth=3, random_state=0)
tree.fit(X, y)
```

```python
# visualize tree

plt.Figure(figsize=(10,6))
plot_tree(tree, filled=True, feature_names=X.columns, class_names=tree.classes_)
plt.title("Fruit Quality Decision Tree")
plt.show()
```

```python
# test prediction 
test_sample = [[80, 6, 6.4]]
pred = tree.predict(test_sample)
print("predicted Quality:", pred[0])
```
